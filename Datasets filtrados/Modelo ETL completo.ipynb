Carregando a base do segundo ano de análise

```python
#Antes de 2015
matr_2014 = pd.read_csv('MATRICULA_SUL_2014.CSV', sep='|', usecols=
                                ['FK_COD_ALUNO', 'FK_COD_ETAPA_ENSINO'])
```

```python
matr_2014.rename(columns={'FK_COD_ALUNO': 'CO_PESSOA_FISICA',
                        'FK_COD_ETAPA_ENSINO': 'TP_ETAPA_ENSINO'}, inplace=True)
```

```python
#2015 em diante
matr_2017 = pd.read_csv('MATRICULA_SUL_2017.CSV', sep='|', usecols=
                                ['CO_PESSOA_FISICA', 'TP_ETAPA_ENSINO'])
```

Filtrando para somente ensino fundamental

```python
matr_2017 = matr_2017[matr_2017.TP_ETAPA_ENSINO.isin(filtro)]
```

Conferindo a dimensão da base de dados

```python
matr_2017.shape
```

Verificando duplicatas de alunos

```python
matr_2017.sort_values(by='CO_PESSOA_FISICA', ignore_index=True, inplace=True)
matr_2017[matr_2017.duplicated(subset='CO_PESSOA_FISICA', keep=False)]
```

Contagem de duplicatas

```python
matr_2017[matr_2017.duplicated(subset='CO_PESSOA_FISICA')].shape
```

Remoção dos valores duplicados

```python
matr_2017.drop_duplicates(subset=['CO_PESSOA_FISICA'], keep=False, inplace=True, ignore_index=True)
```

Nova verificação de duplicatas

```python
matr_2017[matr_2017.duplicated(subset='CO_PESSOA_FISICA')].shape
```

Recodificando os anos escolares para facilitar a análise posterior

```python
matr_2017['TP_ETAPA_ENSINO'].replace(14, 1, inplace=True) # 1° ano
matr_2017['TP_ETAPA_ENSINO'].replace([4, 15], 2, inplace=True) # 2° ano
matr_2017['TP_ETAPA_ENSINO'].replace([5, 16], 3, inplace=True) # 3° ano
matr_2017['TP_ETAPA_ENSINO'].replace([6, 17], 4, inplace=True) # 4° ano
matr_2017['TP_ETAPA_ENSINO'].replace([7, 18], 5, inplace=True) # 5° ano
matr_2017['TP_ETAPA_ENSINO'].replace([8, 19], 6, inplace=True) # 6° ano
matr_2017['TP_ETAPA_ENSINO'].replace([9, 20], 7, inplace=True) # 7° ano
matr_2017['TP_ETAPA_ENSINO'].replace([10, 21], 8, inplace=True) # 8° ano
matr_2017['TP_ETAPA_ENSINO'].replace([11, 41], 9, inplace=True) # 9° ano
```

Unificando os datasets pelo código do aluno, usando o primeiro dataset como referência e renomeando as colunas com os anos respectivos

```python
matr = pd.merge(matr_2016, matr_2017, how='left', on='CO_PESSOA_FISICA', suffixes=('', '_2017'))
```

Substituindo NA na etapa de ensino do segundo ano para zeros

```python
matr.TP_ETAPA_ENSINO_2017.fillna(value=0, inplace=True)
```

Apresentação dos 100 primeiros dados do dataset unificado

```python
matr.head(100)
```

Dimensão final do dataset (deve ser igual à dimensão do primeiro dataset tratado)

```python
matr.shape
```

Verificando se sobraram duplicatas no dataset unificado

```python
matr.sort_values(by='CO_PESSOA_FISICA', ignore_index=True, inplace=True)
matr[matr.duplicated(subset='CO_PESSOA_FISICA', keep=False)].shape
```

Excluir as bases iniciais para aliviar a memória

```python
del matr_2016
del matr_2017
```

Função lambda para classificação do status de cada aluno (repetente, evadido, aprovado ou erro)

```python
matr['Status'] = matr.apply(lambda x: 'repetente' if (x['TP_ETAPA_ENSINO'] == x['TP_ETAPA_ENSINO_2017']) 
                            else ('aprovado' if x['TP_ETAPA_ENSINO_2017'] == x['TP_ETAPA_ENSINO']+1 
                                  else ('evadido' if x['TP_ETAPA_ENSINO_2017'] == 0 else 'erro')), axis=1)
matr.head()
```

Contagem de dados das categorias criadas acima

```python
matr.value_counts(subset='Status')
```

Valores classificados como erro são inconsistências dos dados que devem removidas da análise

```python
matr[matr.Status=='erro']
```

Removendo as linhas com erros

```python
matr = matr[matr.Status!='erro']
```

Conferindo o resultado final

```python
matr.value_counts(subset='Status')
```

Removendo os dados de 2017 e a coluna de município que não serão mias úteis.

```python
matr.drop(columns=['CO_MUNICIPIO_END', 'TP_ETAPA_ENSINO_2017'], inplace=True)
```

Aplicando one hot encoding ao status dos alunos

```python
matr = pd.get_dummies(matr)
```

Gerando arquivo CSV do resultado

```python
matr.to_csv('evasão_16.csv')
```

Precisamos normalizar os dados e aí podemos usar histogramas para conhecer melhor a distribuição. No caso abaixo por exemplo em termos absolutos temos maior reprovação de brancos, mas quando normalizarmo pode ser que relativo ao total o cenário fique diferente.

```python
g12 = sns.catplot(x="TP_COR_RACA", kind="count", palette="ch:.25", data=evadido_12, aspect=1.5)
g12.set_xticklabels(["Não declarada", "Branca", "Preta", "Parda", "Indígena"])
```
